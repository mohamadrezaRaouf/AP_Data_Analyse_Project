```python
import pandas as pd
import matplotlib.pylab as plt
import datetime
import yfinance as yf
```

```python
ticker_gold = "GC=F"
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime.now()
gold_data = yf.download(ticker_gold, start=start_date, end=end_date)
gold_data.to_csv('gold.csv')
print(gold_data)
```

```python
ticker_oil = "CL=F"
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime.now()
oil_data = yf.download(ticker_oil, start=start_date, end=end_date)
oil_data.to_csv('oil.csv')
print(oil_data)
```

```python
oil_data = oil_data[(oil_data.index >= '2010-01-01') & (oil_data.index <= '2023-07-29')]

oil_data_resampled=oil_data.resample('D').last()

oil_data_resampled=oil_data_resampled.bfill()

oil_data_resampled.to_csv('oil.csv')
oil_data_resampled
```

```python
gold_data = gold_data[(gold_data.index >= '2010-01-01') & (gold_data.index <= '2023-07-29')]

gold_data_resampled=gold_data.resample('D').last()

gold_data_resampled=gold_data_resampled.bfill()

gold_data_resampled.to_csv('gold.csv')
gold_data_resampled
```

<div style="text-align: right">خط "اصلی" قیمت های بسته شدن روزانه  را نشان می دهد. خط "میانگین متحرک ۵۰ روزه" میانگین قیمت بسته شدن را در ۵۰ روز گذشته نشان می دهد، در حالی که خط "میانگین متحرک ۲۰۰ روزه" میانگین قیمت بسته شدن را در ۲۰۰ روز گذشته نشان می دهد.
  با مقایسه قیمت های اصلی با میانگین متحرک می توان موارد زیر را مشاهده کرد</div>


<div style="text-align: right">   شناسایی روند: میانگین های متحرک نوسانات روزانه را هموار می کند و تشخیص روند کلی قیمت طلا را آسان تر می کند. اگر میانگین متحرک ۵۰ روزه به طور مداوم بالاتر از میانگین متحرک ۲۰۰ روزه باشد، نشان دهنده یک روند صعودی است که نشان می دهد قیمت  در حال افزایش بوده است. برعکس، اگر میانگین متحرک ۵۰ روزه به طور مداوم زیر میانگین متحرک ۲۰۰ روزه بماند، نشان دهنده روند نزولی است که نشان می دهد قیمت  در حال کاهش بوده است</div>


<div style="text-align: right"> سطوح حمایت و مقاومت: میانگین های متحرک می توانند به عنوان سطوح حمایت و مقاومت عمل کنند. اگر قیمت از میانگین های متحرک عبور کند، ممکن است نشان دهنده یک سیگنال صعودی باشد که حاکی از حرکت  صعودی قیمت ها است. از سوی دیگر، اگر قیمت از میانگین های متحرک عبور کند، ممکن است نشان دهنده یک سیگنال نزولی باشد که حاکی از حرکت نزولی  قیمت ها است</div>


<div style="text-align: right">متقاطع ها: نمودار همچنین تلاقی های بین میانگین های متحرک را برجسته می کند. زمانی که میانگین متحرک ۵۰ روزه فراتر از میانگین متحرک ۲۰۰ روزه عبور کند، به عنوان "صلیب طلایی" شناخته می شود که یک سیگنال صعودی در نظر گرفته می شود. برعکس، زمانی که میانگین متحرک ۵۰ روزه پایین از میانگین متحرک ۲۰۰ روزه حرکت کند، به عنوان "تصلیب مرگ" شناخته می شود که یک سیگنال نزولی درنظر گرفته می شود</div>

```python
import matplotlib.pyplot as plt

def moving_average(data, window_size1=50,window_size2=200):
    # Compute the moving average
    data['MA'] = data['Close'].rolling(window=window_size1).mean()
    data['MA2'] = data['Close'].rolling(window=window_size2).mean()
    # Plot the original 'Close' prices and the moving average
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Original')
    plt.plot(data.index, data['MA'], label=f'{window_size1}-day Moving Average')
    plt.plot(data.index, data['MA2'], label=f'{window_size2}-day Moving Average')
    plt.title('Close Prices with Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.show()
```

```python
moving_average(gold_data,50,200)
```

```python
moving_average(transformed_gold,50,200)
```

```python
moving_average(normalized_gold,50,200)
```

```python
# Compute the 30-day moving average
gold_data['MA_30'] = gold_data['Close'].rolling(window=30).mean()

# Plot the original 'Close' prices and the 30-day moving average
plt.figure(figsize=(12, 6))
plt.plot(gold_data.index, gold_data['Close'], label='Gold Close')
plt.plot(gold_data.index, gold_data['MA_30'], label='30-day Moving Average')
plt.title('Gold Close Prices with 30-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()

# Identify potential buy or sell signals
sell_signals = gold_data[gold_data['Close'] < gold_data['MA_30']]
buy_signals = gold_data[gold_data['Close'] > gold_data['MA_30']]

# Plot buy signals
plt.scatter(buy_signals.index, buy_signals['Close'], color='green', label='Buy')

# Plot sell signals
plt.scatter(sell_signals.index, sell_signals['Close'], color='red', label='Sell')

plt.legend()
plt.show()
```

<div style="text-align: right"> مولفه 3: این جزء تفاوت بین قیمت های بالا و پایین هر روز را نشان می دهد.برای *های*بار مثبت و برای *لوو*بار منفی دارد. این بدان معنی است که وقتی قیمت بالا بالاتر از حد معمول است، قیمت پایین به سمت پایین تر از حد معمول می رود و بالعکس. این مؤلفه حدود 0.2 درصد از واریانس داده ها را توضیح می دهد </div>


<div style="text-align: right">  مولفه 4: این جزء تفاوت بین قیمت باز و بسته شدن هر روز را نشان می دهد.برای*اوپن*بارگذاری مثبت و برای *کلوز*بارگذاری منفی دارد. این بدان معنی است که وقتی قیمت باز بالاتر از حد معمول است، قیمت بسته به سمت پایین تر از حد معمول می رود و بالعکس. این مؤلفه حدود 0.1 درصد از واریانس داده ها را توضیح می دهد</div>

<div style="text-align: right">مولفه 5: این مؤلفه برخی از تغییرات جزئی را که توسط مؤلفه های قبلی توضیح داده نشده اند، ثبت می کند.دارای بارگذاری مثبت برای *والیوم* و بارگذاری منفی برای *ادجاست کلوز*میباشد. این بدان معنی است که وقتی حجم بالاتر از حد معمول است، قیمت بسته تعدیل شده کمتر از حد معمول است و بالعکس. این مؤلفه حدود 0.03 درصد از واریانس داده ها را توضیح می دهد</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
 # Use data from normalized_USD_EURO dataframe
X = normalized_USD_EURO.values
y = normalized_USD_EURO.index
X_centered = X - np.mean(X, axis=0)
covariance_matrix = np.cov(X_centered.T)
eigenvalues, eigenvectors = linalg.eig(covariance_matrix)
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues_sorted = eigenvalues[sorted_indices]
eigenvectors_sorted = eigenvectors[:, sorted_indices]
print(eigenvalues_sorted)
print(eigenvectors_sorted)
num_components = 2
selected_eigenvectors = eigenvectors_sorted[:, :num_components]
X_transformed = np.dot(X_centered, selected_eigenvectors)

plt.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y, cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - USD to EURO Exchange Rate')
plt.colorbar(label='Classes')

# Plot the eigenvectors
origin = np.mean(X_transformed, axis=0)
for i in range(num_components):
    plt.arrow(origin[0], origin[1], selected_eigenvectors[0, i], selected_eigenvectors[1, i], color='r',
              width=0.02, head_width=0.05)

# Annotate the eigenvalues
for i in range(num_components):
    plt.text(selected_eigenvectors[0, i] + origin[0], selected_eigenvectors[1, i] + origin[1],
             f"λ{i+1} = {eigenvalues_sorted[i]:.2f}", fontsize=8)


plt.show()
```
